```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import img_to_array, load_img
```

```python
img_height, img_width = 64, 64
batch_size = 32
epochs = 50
```

```python
def load_images_from_folder(folder):
    images = []
    labels = []
    for filename in os.listdir(folder):
        img = load_img(os.path.join(folder, filename), target_size=(img_height, img_width))
        if img is not None:
            img_array = img_to_array(img) / 255.0
            images.append(img_array)
            if 'Bart' in folder:
                labels.append(0)
            elif 'Homer' in folder:
                labels.append(1)
    return np.array(images), np.array(labels)
```

```python
bart_images, bart_labels = load_images_from_folder(os.path.join('Bart'))
homer_images, homer_labels = load_images_from_folder(os.path.join('Homer'))
```

```python
X = np.concatenate((bart_images, homer_images), axis=0)
y = np.concatenate((bart_labels, homer_labels), axis=0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

train_generator = train_datagen.flow(X_train, y_train, batch_size=batch_size)
```

```python
model = Sequential([
    Flatten(input_shape=(img_height, img_width, 3)),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test))
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test accuracy: {test_accuracy:.4f}')
```
